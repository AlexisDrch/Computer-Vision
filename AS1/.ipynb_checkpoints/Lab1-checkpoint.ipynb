```python
import cv2
import numpy as np
import sys
from skimage.feature import peak_local_max
from matplotlib import pyplot as plt

img = cv2.imread('ps1-input0.png',0)
edges = cv2.Canny(img,100,200,1,7,True)

plt.figure()
plt.imshow(img,cmap='gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.show()

plt.figure()
plt.imshow(edges,cmap='gray')
plt.title('1) a. Canny'), plt.xticks([]), plt.yticks([])
plt.show()


img=cv2.cvtColor(img, cv2.COLOR_GRAY2BGR); 
```

```python
accuracyDrop=2
ticksTheta=3. # from 0.5 to 0.5 in degrees

height, width = edges.shape

diag=np.ceil(np.sqrt((height)**2 + (width)**2))
aMax=int(diag/accuracyDrop)
bMax=int(90*ticksTheta) #from 0 to pi
accu=np.zeros((aMax,bMax))
y_idx, x_idx = np.nonzero(edges) # find all edge (nonzero) pixel indexes

for i in range(len(x_idx)): # cycle through edge points
    x = x_idx[i]
    y = y_idx[i]
    for j in range(0, bMax): # cycle through thetas and calc rho
        rho = int((x * np.cos(np.deg2rad(j/ticksTheta)) +y * np.sin(np.deg2rad(j/ticksTheta))))
        accu[int(rho/accuracyDrop), j] += 1
        

#  find peaks
xy = peak_local_max(accu, min_distance=20,threshold_rel=0.5,exclude_border=0)
x= [item[0] for item in xy]
y= [item[1] for item in xy]

plt.figure()
imgplot = plt.imshow(accu,cmap='gray')
plt.title('Accu with max'), plt.xticks([]), plt.yticks([])
plt.scatter(y,x)
plt.show()

```

```python


lineConstant=10000
for i in range(0,len(x)):
    a=np.cos(np.deg2rad(y[i]/ticksTheta))
    b=np.sin(np.deg2rad(y[i]/ticksTheta))
    x1Image=int(a*x[i]*accuracyDrop-lineConstant*b) #x in hough space is rho
    y1Image=int(b*x[i]*accuracyDrop+lineConstant*a)
    x2Image=int(a*x[i]*accuracyDrop+lineConstant*b)
    y2Image=int(b*x[i]*accuracyDrop-lineConstant*a)
    cv2.line(img,(x1Image,y1Image),(x2Image,y2Image), (0,255,0),2)
    print(' a= '+str(a)+' b= '+str(b)+' x1= '+str(x1Image)+' y1= '+str(y1Image)+' x2= '+str(x2Image)+' y2= '+str(y2Image))
    plt.imshow(img)
    plt.title('Original with lines'), plt.xticks([]), plt.yticks([])
    plt.show()
```
